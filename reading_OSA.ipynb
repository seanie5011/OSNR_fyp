# FYP OSNR: Reading from OSA Data

Imports:

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Reading from a CSV file

We import the CSV file from the following file path:

```python
file_path = 'Data/OSA/OSA_BASIC.CSV'
```

Then we can import this data using ``numpy``:

```python
# the number of lines we need to skip to get to the data
skiprows = 29
# import the data from the file
# note we have to skip certain rows as they contain settings data
# unpack=True automatically transposes the data so we dont have to
x, y = np.loadtxt(file_path, delimiter=',', skiprows=skiprows, unpack=True)
x, y
```

Finally we can plot using ``matplotlib``:

```python
# PLOTTING
fig, ax = plt.subplots(
    1, 1,
    figsize=(16, 9)
)

# Plot the OSA Data
ax.plot(x, y, label='OSA Data')

ax.legend()

ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel('Power (dBm)')
```

If we want to look at the various settings given, we can import them:

```python
# read all lines from file up to skiprows
with open(file_path, 'r') as file:
    all_text = file.readlines()[:skiprows]

all_text
```
